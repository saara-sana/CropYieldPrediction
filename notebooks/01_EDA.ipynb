```python
import seaborn as sns

# Optional: Makes plots look nicer
sns.set_style("whitegrid")
```

```python
plt.figure(figsize=(8,5))

sns.histplot(df["Yield"], bins=30, kde=True)

plt.title("Distribution of Crop Yield")
plt.xlabel("Yield")
plt.ylabel("Frequency")

plt.show()
```

```python
plt.figure(figsize=(12,6))

sns.countplot(
    data=df,
    x="Crop",
    order=df["Crop"].value_counts().index
)

plt.xticks(rotation=90)

plt.title("Number of Records for Each Crop")

plt.show()
```

```python
plt.figure(figsize=(8,5))

sns.histplot(df["Annual_Rainfall"], bins=30, kde=True)

plt.title("Distribution of Annual Rainfall")

plt.show()
```

```python
plt.figure(figsize=(8,6))

sns.scatterplot(
    data=df,
    x="Annual_Rainfall",
    y="Yield"
)

plt.title("Rainfall vs Yield")

plt.show()
```

The dataset contains multiple crop types. Some crops have significantly more records than others, indicating an imbalance in the number of observations across crops.

```python
plt.figure(figsize=(15,6))

sns.countplot(
    data=df,
    x="Crop",
    order=df["Crop"].value_counts().index
)

plt.xticks(rotation=90)
plt.title("Number of Records for Each Crop")
plt.xlabel("Crop")
plt.ylabel("Count")

plt.show()
```

### Observation

- Rice has the highest number of records in the dataset, followed by Maize and Green Gram.
- Several crops have significantly fewer observations.
- This indicates that the dataset is imbalanced with respect to crop types.
- Such imbalance may affect the model's performance on crops with fewer training examples.

```python
plt.figure(figsize=(18,6))

sns.countplot(
    data=df,
    x="Crop",
    order=df["Crop"].value_counts().index
)

plt.xticks(rotation=90, fontsize=8)

plt.title("Number of Records for Each Crop")
plt.tight_layout()

plt.show()
```

```python
plt.figure(figsize=(8,5))

sns.countplot(data=df, x="Season")

plt.title("Number of Records by Season")
plt.xticks(rotation=45)

plt.show()
```

### Observation

- Kharif season has the highest number of records, followed by Rabi and Whole Year.
- Summer, Autumn, and Winter have comparatively fewer observations.
- This indicates an imbalance in the distribution of records across different seasons, which may influence model performance.

```python
plt.figure(figsize=(12,6))

df["State"].value_counts().head(10).plot(kind="bar")

plt.title("Top 10 States by Number of Records")
plt.xlabel("State")
plt.ylabel("Count")

plt.xticks(rotation=45)

plt.show()
```

### Observation

- Karnataka has the highest number of crop records, followed by Andhra Pradesh and West Bengal.
- The top 10 states contribute a significant portion of the dataset.
- Although the number of records varies across states, the distribution among the top states is relatively balanced.
- This diversity helps the model learn agricultural patterns from different geographical regions.

```python
corr = df.corr(numeric_only=True)
corr
```

```python
plt.figure(figsize=(8,6))

sns.heatmap(
    corr,
    annot=True,
    cmap="coolwarm",
    fmt=".2f"
)

plt.title("Correlation Heatmap")

plt.show()
```

## Correlation Heatmap - Observation

- Production has a moderate positive correlation (0.57) with Yield.
- Area is strongly correlated with Fertilizer (0.97) and Pesticide (0.97).
- Fertilizer and Pesticide are also highly correlated (0.95).
- Annual Rainfall has a weak linear correlation with Yield (0.02).
- Crop Year shows almost no linear relationship with the numerical features.
- Strong correlations among Area, Fertilizer, and Pesticide indicate multicollinearity, which should be considered when selecting machine learning models.
